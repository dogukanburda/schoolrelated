```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

def func(x,n0,lam):
    return n0*np.exp(-lam*x)



deger=[93,0.01] #ASSUMPTION FOR CURVE_FIT


y=[35,52,62,69,75,66,70,71,62,83,65,77,62,73,75,93,92,66,64,88,77,64,83,71,67,67,60,68,52,57,57,61,48,55,\
        54,55,49,44,34,49,46,56,60,42,44,47,44,36,41,38,40,37,36,31,28,23,33,36,30,25,28,22,34,21,21,36,16,33,\
       22,35,24,27,20,21,16,29,23,22,25,36,31,18,23,20,19,17,17,19,18,16,25,6,19,21,24,18,28,16,21,10,17,14,18,\
       17,14,18,17,9,14,18,12,18,21,12,17,12,18,18,21,22,20,17,14,18,18,15,24,15,15,22,11,20,17,16,13,23,13,19,\
       16,20,19,11,15,15,15,21,17,15,12,19,23,18,16,22,18,15,11,12,16,18,14,22,10,24,14,14,13,16,11,13,12,17,11,\
       18,17,12,12,18,19,10,15,14,21,20,9,18,19,16,6,17,14,15,17,14,18,10,17,19,10,16,21,21,20,12,17,16,13,13,19,\
       13,13,22,13,11,8,14,16,18,21,19,20,19,15,9,12,15,16,11,14,16,10,16,15,18,15,13,26,12,24,13,17,8,17,10,8,13,\
       12,15,13,14,12,14,7,13,13,15,9,17,11,11,10,16,11,17,14,20,20,12,13,10,13,7,12,15,11,12,8,13,11,15,14,14,\
       14,14,12,12,13,22,15,14,12,11,15,15,12,11,12,12,13,16,10,18,17,16,19,8,7,17,12,14,17,18,9,12,13,10,14,19,\
       11,18,19,7,17,18,15,14,10,10,14,15,17,11,13,20,20,6,14,15,9,14,18,7,11,8,14,13,16,14,9,15,13,8,19,13,10,\
       16,19,10,20,12]

x=[]
dy=[]


for k in range(len(y)):
    x.append(k*6)
    dy.append(np.sqrt(y[k]))


x=np.array(x)
dy=np.array(dy)


c,v=curve_fit(func,x[15:80],y[15:80],deger)
print(c)
plt.figure(dpi=500,figsize=(10,4.8))

plt.errorbar(x,y, yerr=dy, fmt='o', elinewidth=1, capsize=0,label ='Sayımlar',zorder=5);

plt.plot(x,func(x,c[0],c[1]),label ='Fit',zorder=10)
plt.annotate(text="y(x)={c0:.2f}*e^(-{c1:.5f}x)$".format(c0 = c[0], c1 = c[1]),xy=(300,50))

plt.legend(loc ='upper right')
plt.xlabel("t(s)")
plt.ylabel("Sayım")
plt.grid()
plt.title("Protactinium Yarı Ömür")
```
