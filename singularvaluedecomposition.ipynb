```python
import numpy as np
```

 - **U** corresponds to the eigenvectors of $AA^T$ 
 -    **V** corresponds to the eigenvectors of $A^TA$ 
 -    **D** corresponds to the eigenvalues $AA^T$ or $A^TA$ which are the same.


A=\begin{bmatrix} 3 & 5 \\ 8 & 2 \\ 9 & 4 \end{bmatrix}

```python
A = np.array([[3, 5], [8, 2], [9, 4]])
```

$AA^T$ 's eigenvectors

```python
np.linalg.eig(A.dot(A.T))[1]
```

 $A^TA$'s eigenvectors

```python
np.linalg.eig(A.T.dot(A))[1]
```

$AA^T$'s and $A^TA$'s eigen values:

```python
np.linalg.eig(A.dot(A.T))[0],np.linalg.eig(A.T.dot(A))[0]
```

The nonzero singular values of A are the square roots of the eigenvalues of $A^TA$ and $AA^T$

$D=$

```python
np.sqrt(np.linalg.eig(A.dot(A.T))[0]),np.sqrt(np.linalg.eig(A.T.dot(A))[0])
```

```python
D=np.sqrt(np.linalg.eig(A.T.dot(A))[0])
```
