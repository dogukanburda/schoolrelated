## Dedektörün high voltage devresinin optimal çalışma voltajını bulmak için yaptığımız sayımların analizi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("voltage.xlsx")
```

```python
df.drop(columns=["Unnamed: 4","Unnamed: 5"])
```

```python
sns.set_style(style="darkgrid")
```

```python
plt.figure(dpi=300)
plt.ylim((1200,2300))
sns.scatterplot(x=df.mA,y=df["Vout(-)"],hue=df["frequency(HZ)"])
plt.annotate(
    text='35 mA',
    xy=(35, 2120),
    xycoords="data",
    fontsize=13,
    xytext=(30, 25),
    textcoords='offset points',
    arrowprops=dict(arrowstyle='->', color='black'),  # Use color black
    horizontalalignment='center',  # Center horizontally
    verticalalignment='center')

plt.title("Change in output wrt mA")
```
