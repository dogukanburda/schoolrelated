```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
%matplotlib inline


def func(x,i0,mu):
    return i0*np.exp(-mu*x)



deger=[900,1] #ASSUMPTION FOR CURVE_FIT


y=[895,589,506,461,365,323,252,196,143,100,73]
backgroundnoise=16
x=[0,1.74/13.6,3.4/13.6,5.14/13.6,6.9/13.6,
   8.64/13.6,10.3/13.6,13.8/13.6,17.2/13.6,
  20.7/13.6,27.4/13.6]
dy=[]


for k in range(len(y)):
    y[k]=y[k]-backgroundnoise
    dy.append(np.sqrt(y[k]))


x=np.array(x)
dy=np.array(dy)


c,v=curve_fit(func,x,y)
print(c)
#plt.figure(dpi=360)

plt.errorbar(x,y, yerr=dy,  fmt='o', elinewidth=1, capsize=3,label ='Sayımlar',zorder=5);

plt.plot(x,func(x,c[0],c[1]),label ='Fit',zorder=10)
plt.annotate(text="y(x)=800.66428*exp(-1.60298x)",xy=(0.5,500))
plt.scatter(x,y)
plt.legend(loc ='upper right')
plt.xlabel("x(cm)")
plt.ylabel("Sayım")
plt.grid()
plt.title("Gama Işınlarının Absorbsiyonu (logscale)")
#plt.yscale("log")
plt.show()
```
